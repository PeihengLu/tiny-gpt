```python
# load up the dataset txt file
with open('input.txt') as f:
    lines = f.readlines()

text = ''.join(lines)
s = lines[0]

# single character tokenization
chars = sorted(list(set(text)))
char_to_int = dict((c, i) for i, c in enumerate(chars))
int_to_char = dict((i, c) for i, c in enumerate(chars))

# encoder and decoder
encode = lambda s: [char_to_int[char] for char in s]
decode = lambda i: [int_to_char[i] for i in i]
```

```python
# encode and store the data
import torch
data = torch.tensor(encode(text), dtype=torch.long)
```

```python
# train test split
split = 0.9
split_idx = int(len(data) * split)
train_data = data[:split_idx]
test_data = data[split_idx:]
```

```python
# load the trainning data
import torch

torch.manual_seed(1337)
batch_size = 32
block_size = 8

def get_batch(split: str) -> [torch.Tensor, torch.Tensor]:
    if split == 'train':
        data = train_data
    elif split == 'test':
        data = test_data
    else:
        raise ValueError('split must be either train or test')

    # generate batch_size random starting indices for the batch
    ix = torch.randint(0, len(data) - block_size, (batch_size,))
    # get the context and target 
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+1+block_size] for i in ix])

    return x, y

# get one batch for testing
# xb, yb = get_batch('train')

# for b in range(batch_size):
#     for t in range(block_size):
#         context = xb[b, :t+1]
#         target = yb[b, t]
#         print(f'context: {context} -> target: {target}')
```
